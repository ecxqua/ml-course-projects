Ввод и вывод данных

```python
import nltk #естествення обработка языка
from nltk.tokenize import RegexpTokenizer #токенизация
from nltk.probability import FreqDist #частота слов
```

```python
user_text = input()
```

```python
user_text.lower()
```

```python
tok = RegexpTokenizer(r'\w+')
```

```python
words = tok.tokenize(user_text) #разбивка на слова
```

```python
words
```

```python
total_words = len(words)
total_words
```

```python
freq_dist = FreqDist(words)
freq_dist
```

```python
for word, i in freq_dist.items():
  print(f"{word} - {round(i/total_words*100,2)}%")
```

```python
percentage = {word:round(i/total_words*100,2) for word, i in freq_dist.items()}
```
